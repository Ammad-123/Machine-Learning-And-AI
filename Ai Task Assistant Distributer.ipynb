```python
# 2️⃣ Import libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from datetime import datetime
import json
import os
```

```python
# 3️⃣ Initialize model and tokenizer
model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# 4️⃣ Daily usage tracking
USAGE_FILE = "usage.json"

# Load usage data
if os.path.exists(USAGE_FILE):
    with open(USAGE_FILE, "r") as f:
        usage_data = json.load(f)
else:
    usage_data = {}

# Reset daily count if a new day
today = datetime.today().strftime("%Y-%m-%d")
if usage_data.get("date") != today:
    usage_data = {"date": today, "count": 0}

MAX_TASKS_PER_DAY = 40
```

```python
# 5️⃣ Function to break a big task into subtasks
def break_task(task):
    prompt = (
        f"Break down the task '{task}' into 5-10 specific, actionable subtasks. "
        f"Provide a numbered list with clear, concrete steps. Each subtask should be a complete sentence.\n\n"
        f"Task: {task}\n\n"
        f"Subtasks:\n1."
    )

    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)

    outputs = model.generate(
        **inputs,
        max_length=400,           # Increased max length
        num_beams=5,              # More beams for better quality
        no_repeat_ngram_size=3,
        early_stopping=True,
        do_sample=True,           # Enable sampling for variety
        temperature=0.8,          # Control randomness
        top_p=0.9,                # Nucleus sampling
        length_penalty=1.2,       # Encourage longer outputs
        min_length=100            # Ensure minimum output length
    )

    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result

```

```python
import re
# 6️⃣ User interaction
if usage_data["count"] >= MAX_TASKS_PER_DAY:
    print(f"❌ You have reached your daily limit of {MAX_TASKS_PER_DAY} tasks. Come back tomorrow!")
else:
    task = input("Enter your big task: ")
    subtasks_text = break_task(task)
    usage_data["count"] += 1
    print("\n✅ AI Suggested Subtasks:")
     # Split based on numbering like 1., 2., 3., etc.
    print("*"*20)
    print(subtasks_text)
    subtasks_list = re.split(r'\d+\.\s+', subtasks_text)
    # The first element may be empty if text starts with "1."
    for subtask in subtasks_list:
        subtask = subtask.strip()
        if subtask:
            print(subtask)

    print("\n")



    # Save updated usage
    with open(USAGE_FILE, "w") as f:
        json.dump(usage_data, f)
```

```python
# AI Task Breakdown Assistant
# Google Colab Ready

# 1️⃣ Install dependencies
!pip install -q transformers torch sentencepiece

# 2️⃣ Import libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from datetime import datetime
import json
import os

# 3️⃣ Initialize model and tokenizer
model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# 4️⃣ Daily usage tracking
USAGE_FILE = "usage.json"

# Load usage data
if os.path.exists(USAGE_FILE):
    with open(USAGE_FILE, "r") as f:
        usage_data = json.load(f)
else:
    usage_data = {}

# Reset daily count if a new day
today = datetime.today().strftime("%Y-%m-%d")
if usage_data.get("date") != today:
    usage_data = {"date": today, "count": 0}

MAX_TASKS_PER_DAY = 3

# 5️⃣ Function to break a big task into subtasks
def break_task(task):
    prompt = f"Break the following task into smaller, manageable subtasks:\n\nTask: {task}\n\nSubtasks:"
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=200)
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result

# 6️⃣ User interaction
if usage_data["count"] >= MAX_TASKS_PER_DAY:
    print(f"❌ You have reached your daily limit of {MAX_TASKS_PER_DAY} tasks. Come back tomorrow!")
else:
    task = input("Enter your big task: ")
    subtasks = break_task(task)
    usage_data["count"] += 1
    print("\n✅ AI Suggested Subtasks:")
    print(subtasks)

    # Save updated usage
    with open(USAGE_FILE, "w") as f:
        json.dump(usage_data, f)
```
